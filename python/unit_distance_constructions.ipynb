# Unit-distance constructions

This notebook is the exploratory companion to `unit_distance_constructions.py`.
It regenerates the visual used in the Beamer slides on the unit distance conjecture.

- Top row: classical Erdos integer-grid mechanism.
- Bottom row: schematic norm-one translation strategy evoking the OpenAI/Sawin/Emmerich certificates.

The figure is pedagogical; it is not the full arithmetic certificate.


```python
from pathlib import Path
import sys

repo_root = Path.cwd().parent if Path.cwd().name == "python" else Path.cwd()
sys.path.insert(0, str(repo_root / "python"))

from unit_distance_constructions import make_figure
```

```python
for outdir in sorted((repo_root / "slides").glob("*/assets/generated")):
    language = "fr" if outdir.parents[1].name == "fr" else "en"
    fig = make_figure(language=language)
    outdir.mkdir(parents=True, exist_ok=True)
    fig.savefig(outdir / "unit_distance_series.png", dpi=260, bbox_inches="tight", pad_inches=0.02)
    fig.savefig(outdir / "unit_distance_series.pdf", bbox_inches="tight", pad_inches=0.02)
    print(f"saved {outdir / 'unit_distance_series.png'}")
    print(f"saved {outdir / 'unit_distance_series.pdf'}")
fig
```
